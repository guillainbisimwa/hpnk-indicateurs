```python
import pandas as pd
import io
from datetime import datetime
```

```python
df1 = pd.read_csv("dataset/orl.csv", sep=";")
```

```python
df1 = df1.fillna("") 
```

```python
row_temp = []
Nom = ''
Postnom = ''
Prénom = ''
time = 0
for index, row in df1.iterrows():
    if time == 0:
        Nom = row.Nom
        Postnom = row.Postnom
        Prénom = row.Prénom
    if row.Nom == Nom and row.Postnom == Postnom and row.Prénom == Prénom and time == 1:
        df1.at[index, 'Date'] = ""
        df1.at[index, 'Nom'] = ""
        df1.at[index, 'Postnom'] = ""
        df1.at[index, 'Prénom'] = ""
        df1.at[index, 'Sexe'] = ""
        df1.at[index, 'Age'] = -1
        df1.at[index, 'Plaintes'] = ""
        df1.at[index, 'Diagnostic final'] = ""
        df1.at[index, 'Diagnostic de présemption'] = ""
        df1.at[index, 'Examens labo'] = ""
    else:
        Nom = row.Nom
        Postnom = row.Postnom
        Prénom = row.Prénom
    time = 1
```

```python
df1 = df1.replace(-1, '')
df1.head(10)
```

```python
# datetime object containing current date and time
now = datetime.now()
# dd/mm/YY H:M:S
dt_string = now.strftime("%d-%m-%Y-%H-%M-%S-")
df1.to_csv(r'export/'+dt_string+'export.csv')
```
